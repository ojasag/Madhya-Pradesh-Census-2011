# Data Visualizition

```python
plt.subplots(figsize = (15,8))
ct = madhyapradesh_census.groupby('District name')['Population'].sum()
ax = ct.sort_values(ascending=False).plot(kind="bar")

plt.show()
```

Indore is the most populated district in Madhya Pradesh followed by Jabalpur & Sagar.

```python
IRPdf  = madhyapradesh_census[['Hindus','Muslims','Christians','Sikhs','Buddhists','Jains','Others_Religions']].sum().head(3)
IRPdfg = IRPdf.plot(kind="pie",radius=2,autopct='%1.1f%%', startangle=90)
IRPdfg.set_title("Religions in Madhya Pradesh",loc="center",fontdict={'fontsize':20},position=(0.5,1.55))
```

Hinduism is the most followed religion in Madhya Pradesh followed by Muslims.
